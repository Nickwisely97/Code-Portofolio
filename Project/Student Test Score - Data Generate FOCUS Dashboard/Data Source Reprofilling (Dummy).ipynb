```python
# EthnicGroup -> City
def city_renaming(x):
    if x.upper() == 'GROUP A':
        return 'Medan'
    elif x.upper() == 'GROUP B':
        return 'Bali'
    elif x.upper() == 'GROUP C': 
        return 'Jakarta'    
    elif x.upper() == 'GROUP D':
        return 'Surabaya'
    elif x.upper() == 'GROUP E':
        return 'Semarang'
    else:
        return 'Not Found'

df['City'] = df['EthnicGroup'].apply(city_renaming)
df.head()
```

```python
df['ParentEduc'].unique()
```

```python
for deg in df['ParentEduc'].unique():
    print(f'''elif x.upper() == "{deg.upper()}":
    return ''')
```

```python
# ParentEduc -> EducLevel
def education_level(x):
    if x.upper() == "SOME COLLEGE":
        return 'Setara SMA'
    elif x.upper() == "HIGH SCHOOL":
        return 'Setara SMP'
    elif x.upper() == "SOME HIGH SCHOOL":
        return 'Diploma'
    elif x.upper() == "ASSOCIATE'S DEGREE":
        return 'S1'
    elif x.upper() == "BACHELOR'S DEGREE":
        return 'S2'
    elif x.upper() == "MASTER'S DEGREE":
        return 'S3'
    else:
        return 'Not Found'

df['EducLevel'] = df['ParentEduc'].apply(education_level)
df.head()
```

```python
for deg in df['LunchType'].unique():
    print(f'''  elif x.upper() == "{deg.upper()}":
        return ''')
```

```python
# LunchType -> PackageType
def package_type(x):
    if x.upper() == "STANDARD":
        return 'Group Package'
    elif x.upper() == "FREE/REDUCED":
        return 'Individual Package'
    else:
        return 'Not Found'

df['PackageType'] = df['LunchType'].apply(package_type)
df.head()
```

```python
df['StudentID'] = 250400000 + df.index + 1
df.tail()
```

```python
df.rename(columns={
    'MathScore': 'Score_Module_01', 
    'ReadingScore': 'Score_Module_02',
    'WritingScore': 'Score_Module_03'
    },inplace=True)

df.head()
```

```python
final_columns = ['StudentID', 'Gender', 'City', 'EducLevel', 'PackageType', 'TestPrep', 'Score_Module_01', 'Score_Module_02', 'Score_Module_03']

final_df = df[final_columns]
```
